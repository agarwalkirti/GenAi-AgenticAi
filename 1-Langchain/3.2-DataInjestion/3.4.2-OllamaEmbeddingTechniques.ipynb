```python
#ollama supports embedding models, making it possible to build retrieval augmented generation(RAG) applications that 
# combine text prompts with existing documents or other data.

#from langchain_community.embeddings import OllamaEmbeddings
from langchain_ollama import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model = 'gemma:2b') # by default it uses llama2 model
)
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is the first letter of Greek alphabet",
        "Beta is the first letter of Greek alphabet",
    ]
)
```

```python
len(r1[0]) 
#gemma model creates vectors of dimension 2048
```

```python
query = embeddings.embed_query("What is the second letter of Greek alphabet")
    
query
```

```python
#other embedding models 
embeddings = OllamaEmbeddings(model = 'mxbai-embed-large') 
text = "This is a test document."
query_result = embeddings.embed_query(text)
query_result
```
